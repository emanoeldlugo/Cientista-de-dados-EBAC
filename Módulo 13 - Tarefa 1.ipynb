```python
reg = smf.ols('''np.log(renda) ~ sexo + posse_de_veiculo + posse_de_imovel + C(qtd_filhos, Treatment(0)) + C(educacao, Treatment("Secundário")) + C(estado_civil, Treatment("Casado"))
              + C(tipo_residencia, Treatment("Casa")) + C(qt_pessoas_residencia, Treatment(2))''' , data = df).fit()
reg.summary()
```
