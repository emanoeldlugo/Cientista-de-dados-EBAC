#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
lista_indicadores = ['Close', 'Open', 'High', 'Low']

ativo = ['B3SA3.SA']*len(lista_indicadores)

colunas = list(zip(lista_indicadores, ativo))
````

````python
b3sa = df.loc['23/04/2025':, colunas]
b3sa.columns = lista_indicadores
````

````python
graph = {
    'x': b3sa.index,
    'open': b3sa['Open'],
    'close': b3sa['Close'],
    'high': b3sa['High'],
    'low': b3sa['Low'],
    'type': 'candlestick',
    'name': 'B3SA.SA',
    'showlegend': True
}
````

````python
layout = Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

grafico = graph_objects.Figure(data=[graph], 
                               ##layout=layout, 
                               layout_title='Gráfico Candlestick B3SA3.SA'
                               )

grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGray')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGray')

grafico
````
