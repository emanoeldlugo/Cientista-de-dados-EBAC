2- Calcule essa √°rea de forma anal√≠tica.

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
warnings.filterwarnings("ignore")

# Fun√ß√£o f(x) = x^3
def f(x):
    return x**3

def calcula_area(a, b, func, frame_num):
    num_bins = frame_num + 1
    delta = (b - a) / num_bins
    x_cols = np.linspace(a, b - delta, num_bins) + delta / 2
    y_cols = func(x_cols)
    return sum(delta * y_cols)

def animate(n, a=1, b=2, func=f, verdadeiro_valor=3.75):
    plt.cla()
    ax = plt.axes(xlim=(a, b), ylim=(0, 10))
    ax.ticklabel_format(style='plain', axis='both')  # üëà aqui √© o que resolve seu problema

    # Plota a fun√ß√£o
    x = np.linspace(a, b, 300)
    y = func(x)
    ax.plot(x, y, 'r-', label='f(x) = x^3')

    # Ret√¢ngulos
    num_bins = n + 1
    delta = (b - a) / num_bins
    x_cols = np.linspace(a, b - delta, num_bins) + delta / 2
    y_cols = func(x_cols)
    bar = ax.bar(x_cols, y_cols, alpha=0.5, width=delta, edgecolor='blue')

    # √Årea e erro
    soma_riemann = calcula_area(a, b, func, n)
    erro = abs(verdadeiro_valor - soma_riemann)
    ax.set_title(f'Soma de Riemann = {soma_riemann:.4f} | Erro = {erro:.4f} | Retangulos = {num_bins}')
    ax.legend()
    return bar

animate(49)
plt.show()
```

```python
import sympy as sp

x = sp.symbols('x')
f = x**3

# Calcula integral definida de 1 at√© 2
area = sp.integrate(f, (x, 1, 2))

print('√Årea exata (fracion√°ria):', area)
print('√Årea exata (decimal):', area.evalf())
```
