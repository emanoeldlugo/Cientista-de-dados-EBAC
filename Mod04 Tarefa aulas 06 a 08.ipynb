````python
link_1 = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl'
link_2 = 'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf=SP&tipo_interessado=Estado'
````

````python
dados_geral = requests.get(link_1)
dados_geral.status_code
````

````python
#1) Seu código aqui
def consulta_pvl(uf, interessado):
    r = requests.get(link_1)
    dados = r.json()
    df = pd.DataFrame(dados['items'])
    df.head(5)
    df = df.loc[df['uf'] == uf]
    df = df.loc[df['tipo_interessado'] == interessado]
    df_def = df
    if len(df_def) > 0:
        print('DataFrame criado com sucesso!')
    else:
        print('Dados inconsistentes, favor repetir o processo!')
    return df_def
````

````python
df_novo = consulta_pvl(uf=input('UF: '), interessado=input('Estado ou Município: '))
````
