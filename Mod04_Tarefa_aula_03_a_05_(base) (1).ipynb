## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui
df_mat['acerto'] = df_mat['Qt_acertos'] / 24
df_por['acerto'] = df_por['Qt_acertos'] / 18
df_geo['acerto'] = df_geo['Qt_acertos'] / 8
df_ing['acerto'] = df_ing['Qt_acertos'] / 8
df_his['acerto'] = df_his['Qt_acertos'] / 8
df_fis['acerto'] = df_fis['Qt_acertos'] / 12
df_qui['acerto'] = df_qui['Qt_acertos'] / 12

df_alunos = pd.DataFrame(data=[df_mat['acerto'], df_por['acerto'], df_geo['acerto'], df_ing['acerto'], df_his['acerto'],
                              df_fis['acerto'], df_qui['acerto']], index=['mat', 'por', 'geo', 'ing', 'his', 'fis', 'qui']).T
df_alunos
````

````python
# 2) Seu código aqui
soma_total = (df_mat['Qt_acertos'] + df_por['Qt_acertos'] + df_geo['Qt_acertos'] + df_ing['Qt_acertos'] + 
              df_his['Qt_acertos'] + df_fis['Qt_acertos'] + df_qui['Qt_acertos'])
soma_total = pd.DataFrame(soma_total)
soma_total
````

````python
# 3) Seu código aqui
questoes_total = 24 + 18 + 8 + 8 + 8 + 12 + 12
soma_total['acertos_final'] = soma_total['Qt_acertos'] / questoes_total
soma_total
````

````python
soma_total['fl_aprovado'] = 0
````

````python
# 4) Seu código aqui
soma_total.loc[soma_total['acertos_final'] > 0.45].shape[0]
````

## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# seu código aqui
df_alunos['flag_desqualificado'] = 0
for linha in df_alunos.index:
    if df_alunos['mat'].iloc[linha] == 0 or df_alunos['fis'].iloc[linha] == 0 or df_alunos['qui'].iloc[linha] == 0:
        df_alunos.at[linha, 'flag_desqualificado'] = 1
````

````python
df_alunos['flag_desqualificado'].head(10)
````

````python
## Novo DF Alunos desclassificados
df_alunos_desclassificados = df_alunos.loc[df_alunos['flag_desqualificado'] == 1]
df_alunos_desclassificados
````

````python
## Quantidade de alunos desclassificados
df_alunos_desclassificados.shape
````

````python
## Média nas matérias história e geografia
media_his = df_alunos_desclassificados['his'].mean()
media_geo = df_alunos_desclassificados['geo'].mean()
print(f'A média dos alunos em história foi de {round(media_his*100,2)} e em geografia foi de {round(media_geo*100,2)}!')
````

````python
## Novo DF para segunda_fase
df_segunda_fase = df_alunos.loc[df_alunos['flag_desqualificado'] == 0]
df_segunda_fase
````
